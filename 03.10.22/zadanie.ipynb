{sin(x) = e^y(x) y'(x), y(0) = 0}

{y'(x) = 1/2 i e^(-i x - y(x)) - 1/2 i e^(i x - y(x)), y(0) = 0}

{y'[x] == I/2 E^(-I x - y[x]) - I/2 E^(I x - y[x]), y[0] == 0}

y(x) = log(2 - cos(x))

```python
import numpy as np
from scipy import integrate
from matplotlib import pyplot as plt

def f(y, x):
    return np.exp(-y)*np.sin(x)

xi = np.linspace(0, 1, 10)
y0 = 0
sol = integrate.odeint(f, y0, xi)






rsol = np.log(2-np.cos(xi))
plt.plot(xi, rsol,
            linestyle='-',
            color='black',
            linewidth=1,
            markersize=3.0,
            label='real solution')

plt.plot(xi, sol,
            linestyle='--',
            color='red',
            linewidth=1,
            markersize=3.0,
            label='solution')
# панель для рисования графика скорости
plt.title('решение оду')
# график x(t) - зависимость нулевого столбца вектора # решения sol от ti
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
er = rsol.reshape(10,1)-sol
plt.plot(xi, er,
            linestyle='-',
            color='blue',
            linewidth=1,
            markersize=3.0,
            label='error')
# панель для рисования графика скорости
plt.title('Относительная ошибка')
# график x(t) - зависимость нулевого столбца вектора # решения sol от ti
plt.xlabel('x')
plt.ylabel('y')
plt.show()


```

```python
def f(x):
  return x**3 / (x**8 -  2)
dx = 0.0155
x = np.arange(0, 1, dx)
fy = f(x)
rs = np.log(np.abs(1-np.sqrt(2))/(1+np.sqrt(2)))/(np.power(2,3.5))
rs
np.size(x)
```

```python
iquad = integrate.quad(f, 0, 1.0)
print('Погрешность quad ', np.abs(iquad[0]-rs))
```

```python
iromb = integrate.romb(fy,dx)
print('Погрешность romb ', np.abs(np.sum(iromb)-rs))
```

```python
fsimps = f(x)
isimps = integrate.simps(fy,x)
print('Погрешность simps ', np.abs(isimps-rs))
```

```python
ftrapz = f(x)
itrapz = integrate.trapz(fy,x)
print('Погрешность trpaz ', np.abs(itrapz-rs))
```
