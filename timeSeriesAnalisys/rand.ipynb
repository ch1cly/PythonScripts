```python
import numpy as np
import matplotlib.pyplot as plt


def CDF(x, lambd=1):
    return 1 - np.exp(-lambd * x)

def inverse_CDF(y, lambd=1):
    return -1/lambd * np.log(1 - y)

num_samples = 1000

rand_nums = np.random.rand(num_samples)
random_variables = inverse_CDF(rand_nums)



plt.plot(rand_nums, random_variables, 'o')
plt.xlabel('Random Variable')
plt.ylabel('Frequency')
plt.title('Simulated Random Variable')
plt.grid(True)
plt.show()

plt.hist(random_variables, bins='auto')
plt.xlabel('Random Variable')
plt.ylabel('Frequency')
plt.title('Simulated Random Variable')
plt.grid(True)
plt.show()


```

```python
rand_nums = np.random.rand(5)
print(rand_nums)
print(CDF(inverse_CDF(rand_nums)))

rand_nums = np.random.rand(1000)
np.allclose(rand_nums , CDF(inverse_CDF(rand_nums)))
```
