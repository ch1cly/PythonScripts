```python
import numpy as np
import matplotlib.pyplot as plt


def CDF(x, lambd=1):
    return 1 - np.exp(-lambd * x)

def inverse_CDF(y, lambd=1):
    return -1/lambd * np.log(1 - y)

num_samples = 1000

rand_nums = np.random.rand(num_samples)
random_variables = inverse_CDF(rand_nums)



plt.plot(rand_nums, random_variables, 'o')
plt.xlabel('Random Variable')
plt.ylabel('Frequency')
plt.title('Simulated Random Variable')
plt.grid(True)
plt.show()

plt.hist(random_variables, bins='auto')
plt.xlabel('Random Variable')
plt.ylabel('Frequency')
plt.title('Simulated Random Variable')
plt.grid(True)
plt.show()


```

```python
rand_nums = np.random.rand(5)
print(rand_nums)
print(CDF(inverse_CDF(rand_nums)))

rand_nums = np.random.rand(1000)
np.allclose(rand_nums , CDF(inverse_CDF(rand_nums)))
```

Для доказательства данного утверждения, что для моделирования случайной величины можно использовать функцию обратную к функции распределения, нужно показать, что функция G(rand()) имеет такое же распределение, как и исходная случайная величина.

Пусть X - исходная случайная величина, F(x) - ее функция распределения, G(y) - функция обратная к F, а Y = rand(). Наша задача - показать, что G(Y) имеет то же распределение, что и X.

Так как Y распределено равномерно на [0,1], то его функция распределения F_Y(y) равна y на интервале [0,1]. Используя утверждение о том, что F(G(y)) = y, получаем:

F_G(Y)(y) = P(G(Y) ≤ y) = P(F(G(Y)) ≤ F(y)) = P(Y ≤ F(y)) = F_Y(F(y)) = F(y)

Таким образом, F_G(Y)(y) = F(y), что означает, что функция распределения G(Y) совпадает с функцией распределения X.
