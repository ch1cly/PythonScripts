```python
import yfinance as yf

df = yf.download('AAPL', start='2021-01-01', end='2021-12-31')
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df.plot(title="DataFrame Plot")
```

```python
plt.plot(df.index,df.Open,df.Close)
```

```python
from matplotlib.pyplot import figure

figure(figsize=(18, 6), dpi=80)
plt.plot(df.index,df.Open,df.Close)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.suptitle('Horizontally stacked subplots')
ax1.plot(df.index, df.Open)
ax2.plot(df.index, df.Close)
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1)
fig.suptitle('vertical stacked subplots')
ax1.plot(df.index, df.Open,'go-',c='green')
ax2.plot(df.index, df.Close,'rs',color='red',markersize=0.1)
```

```python
figure(figsize=(15, 6), dpi=80)
plt.plot(df.index,df.Open,'-.',c='blue')
plt.plot(df.index,df.Close,'.',c='red')
plt.legend(['Open','Close'])
plt.title("Title text")
plt.xlabel("Sample x")
plt.ylabel("Sample y")

plt.grid()
plt.show()
```

```python
figure(figsize=(15, 6), dpi=80)
plt.plot(df.index,df.Open,'-.',c='blue')
plt.plot(df.index,df.Close,'.',c='red')
plt.legend(['Open','Close'],markerscale=3,fontsize=20,loc='right',ncol=2,shadow=True,title='legend title')
plt.title("Title text")
plt.xlabel("Sample x")
plt.ylabel("Sample y")

plt.grid()
plt.show()
```

```python
figure(figsize=(15, 6), dpi=80)
plt.step(df.index[:30],df.Open[:30], where='mid', label='post',c='red')
plt.step(df.index[:30],df.Close[:30], where='mid', label='mid',c='blue')
plt.plot(df.index[:30],df.Open[:30], 'o--', color='grey', alpha=0.3)
plt.plot(df.index[:30],df.Close[:30], 'o--', color='black', alpha=0.9)
plt.legend(['Open','Close'])

plt.show
```
